```python
import ast

# Define a function to parse a string into a list
def parse_production_companies(companies):
    if isinstance(companies, str):
        try:
            # Convert a string to a list using ast.literal_eval
            return ast.literal_eval(companies)
        except ValueError:
            return None
    return companies
```

```python
# Convert the strings in the production_companies column to a list
film_train['production_companies'] = film_train['production_companies'].apply(parse_production_companies)
```

```python
def extract_main_company(companies):
    if isinstance(companies, list) and len(companies) > 0 and isinstance(companies[0], dict):
        return companies[0].get('name')
    return None

# Create the main_companies column to extract the main company name
film_train['main_companies'] = film_train['production_companies'].apply(extract_main_company)

print(film_train[['production_companies', 'main_companies']].head(20))
```

```python
Y_train = film_train['revenue']
```

```python
columns_to_drop=["genres","spoken_languages","production_countries","production_companies","id","belongs_to_collection","genres","homepage","imdb_id","status","original_title","overview","poster_path","production_companies","production_countries","spoken_languages","tagline","title","Keywords","cast","crew"]
film_train.drop(columns_to_drop,axis=1,inplace=True)
film_test.drop(columns_to_drop,axis=1,inplace=True)
```

```python
X_train = film_train[['budget', 'original_language', 'popularity', 'runtime','revenue']]
```

```python
X_test = film_test[['budget', 'original_language', 'popularity', 'runtime']]
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# Encode the training set
encoded_language_train = encoder.fit_transform(X_train[['original_language']])

# Encoding the test set
encoded_language_test = encoder.transform(X_test[['original_language']])


# Merge the encoded features back into the original data
X_train_encoded = np.hstack([X_train[['budget', 'popularity', 'runtime']], encoded_language_train])
X_test_encoded = np.hstack([X_test[['budget', 'popularity', 'runtime']], encoded_language_test])
```

```python
X_train_encoded
```

```python
X_test_encoded
```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor

# Create an interpolator
imputer = SimpleImputer(strategy='mean')

# Create a random forest regression model
model = RandomForestRegressor()

# Create a pipeline
pipeline = Pipeline(steps=[
    ('imputer', imputer),  # Add interpolation steps
    ('regressor', model)   # Add regression step
])
```

```python
Y_train
```

```python
pipeline.fit(X_train_encoded, Y_train)
```

```python
Prediction = pipeline.predict(X_test_encoded)
```

```python
# Assuming 'ids' is the range of IDs starting from 3001
ids = range(3001, 3001 + len(Prediction))

#prediction
Predictions = pd.DataFrame({"id": ids, "revenue": Prediction})

print(Predictions)
```
